# MATPLOTLIB LEARNING

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(1,10)
y=np.arange(11,20)
a=np.arange(40,50)
b=np.arange(50,60)
```

```python
plt.scatter(x,y,c=None)
plt.show()
```

```python
plt.scatter(x,y,c=None)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2d')
```

```python
y=x*x
```

```python
plt.plot(x,y,'ro')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'b*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('graph in 2d')
plt.show()
```

```python
# creating subplot
plt.subplot(2,2,1)
plt.plot(a,b,'r--')
plt.subplot(2,2,2)
plt.plot(a,b,'b*-')
plt.subplot(2,2,3)
plt.plot(a,b,'gs')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y,"r")
plt.title('sine wave 2D')
plt.show()
```

```python
y_cos=np.cos(x)
plt.plot(x,y_cos)
plt.title('cosine wave 2D')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.title('sine wave 2D')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r*')
plt.title('cosine wave 2D')
```

```python
# bar plot
x=[2,3,4,5]
y=[3,4,5,6]
x1=[2,9,3,5]
y1=[5,8,6,4]
plt.bar(x,y)
plt.bar(x1,y1,color='r')
plt.show()
```

# Histogram

```python
a=np.array([22,44,45,34,65,67,65,76,44])
plt.hist(a)
plt.title('histogram')
plt.show()
```

# Piechart

```python
 labels='mon','tue','wed','thurs'
 sizes = [234,566,223,443]
 colors=['yellow','red','green','blue']
 explode=(0.2,0,0,0)
 plt.pie(sizes,explode=explode,labels=labels, colors=colors,autopct='%1.1f%%', shadow=True)
 plt.axis('equal')   
plt.show()
    
```
